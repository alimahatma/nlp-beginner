```python
#masukkan packge NLTK
import nltk

#data sampel yang digunakan untuk latihan

text = """Hello Mr.Ali, how are you doaing today?
        The weater is great, and the sky is pinlish-blue. 
        You shouldn't eat cardboard.
        """
print(text)
```

#TOKENISASI
tokenisasi adalah langkah pertama dalam analisis text. Tokenisasi merupakan proses memecah paragraf text menjadi potongan-potongan kecil seperti kata atau kalimat disebut tokenization;

```python
#Word Tokenization
from nltk.tokenize import sent_tokenize
tokenized_text = sent_tokenize(text)
print(tokenized_text)
```

```python
#Frequency Distribution
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(text)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
fdsit = FreqDist(tokenized_word)
print(fdsit)
print(fdsit.most_common(2))
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
fdsit.plot(10, cumulative=False)
plt.show()
```

```python
#stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

#STOP WORD
stop word dianggap sebagai derau dalam text. Text dapat berisi kata-kata seperti: is, am, ate, this, a, an, the, dll. di NLTK untuk menghapus stopword kita perlu daftar stopwords dan memfilter daftar token dari kata kata ini:

```python
#remove stopwords
filter_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filter_sent.append(w)
print("Tokenized Sentence:\n ", tokenized_text)
print("\nFilter Sentence : \n", filter_sent)
```

#Lexicon Normalization:

Lexsion normalize mempertimbangkan jenis kebisingan dalam text, menghubungkan kata menguranginya menjadi kata umum. mengurangi bentuk-bentuk kata yang terkait secara derivatif menjadi kata dasar yang umum.

#Stemming

stemming merupakan proses normalisasi ligistik, yang mereduksi kata-kata menjadi kata dasar atau memotong afiks derivasional. misalnya koneksi, terhubung, menghubungkan kata mengurangi menjadi kata umum

```python
#steaming
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

steam_word = []
for w in filter_sent:
    steam_word.append(ps.stem(w))
#

print("Filter Sentence :\n ", filter_sent)
print("--------------------------------------------")
print("Steammed Sentence : \n", steam_word)
```

#LEMMATZATION

lematization mereduksi kata menjadi kata dasar, lematization merubah kata dasar dengan menggunakan analisis kosa kata dan morfologi. Lematization biasanya lebih canggih daripada steaming. Stemmer bekerja pada kata individual tanpa pengetahuan tentang konteksnya. Misalnya, Kata "better" memiliki "good" sebagai lammanya. Hal iini akan hilang dengan membendung karena membutuhkan kamus pencarian.

```python
#lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer
stem = PorterStemmer()

word= "ALbert Enstain isn't understand about pysich"

print("Lematized Word : ", lem.lemmatize(word, "v") )
print("Stemmed Word : ", stem.stem(word))
```

#POS TAGGING

target utama dari penandaan Part-of-speech (POST) adalah untuk mengidentifikasi grup gramatikal dari kata yang diberikan. apakah itu kata benda, prnoun, adjektif, verb, penandaan POS mencari hubungan dalam kalimat dan memberikan tag yang sesuai untuk kata tersebut   

```python
sent = stem.stem(word)
tokens = nltk.word_tokenize(sent)
print(tokens)
```

```python
result_POS = nltk.pos_tag(tokens)
print(result_POS)
```

# Performing Sentiment Analysis using Text Classification

```python
import pandas as pd
data = pd.read_csv('train.tsv', sep='\t')
data.head()
```

```python
data.info()
```

```python
data.Sentiment.value_counts()
```

```python
Sentiment_count=data.groupby('Sentiment').count()
plt.bar(Sentiment_count.index.values, Sentiment_count['Phrase'])
plt.xlabel("Review Sentiments")
plt.ylabel("Number of Sentiments")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(
    lowercase=True, 
    stop_words="english",
    ngram_range=(1,1), tokenizer=token.tokenize
)
text_count = cv.fit_transform(data['Phrase'])
text_count
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_count, data['Sentiment'],
    test_size=0.3,
    random_state=1
)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

clf = MultinomialNB().fit(X_train, y_train)
predictions = clf.predict(X_test)

print("MultinomialND: ", metrics.accuracy_score(y_test, predictions))
```
